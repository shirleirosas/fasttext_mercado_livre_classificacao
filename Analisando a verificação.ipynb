```python
import matplotlib.pyplot as plt
import pandas as pd 
from unidecode import unidecode
import re
```

```python
df_verificacao = pd.read_csv("frameVerificacao- comparacao.csv")
```

```python
df_verificacao_or_procen=df_verificacao.sort_values(by=["Porcentagem_Massa_teste"], ascending=False)
df_verificacao_or_procen.head()
```

```python
plt.plot( 'Categoria', 'Porcentagem_Massa_teste', data=df_verificacao,  color='green',linestyle='-')
plt.xticks(rotation='vertical')
plt.xlabel('Categoria')
plt.ylabel('Porcentagem')
plt.rcParams["figure.figsize"] = (20,3)
frame1 = plt.gca()
frame1.axes.xaxis.set_ticklabels([])
plt.show()
```

```python
plt.plot( 'Categoria', 'PorcentagemTotal', data=df_verificacao, color='skyblue',linestyle='-')
plt.xticks(rotation='vertical')
plt.xlabel('Categoria')
plt.ylabel('Porcentagem')
plt.rcParams["figure.figsize"] = (20,3)
frame1 = plt.gca()
frame1.axes.xaxis.set_ticklabels([])

plt.show()         
```

```python
plt.plot( 'Categoria', 'PorcentagemTotal', data=df_verificacao.head(7), marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'Categoria','Porcentagem_Massa_teste', data=df_verificacao.head(7), marker='x', color='olive', linewidth=2)
plt.title('Relação entre a porcentagem errada * porcentagem da amostra')
plt.xticks(rotation='vertical')
plt.xlabel('Categoria')
plt.ylabel('Porcentagem')
plt.rcParams["figure.figsize"] = (20,3)
plt.legend()
plt.grid(True)
plt.show()
```
