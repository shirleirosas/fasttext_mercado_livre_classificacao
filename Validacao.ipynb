```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("resultado/resultadoFinal-massaClassificada.csv")
df.head()
```

```python
df_diferente = df[df.category!= df.predicao]
df_diferente.head(20)
```

```python
tamanho_df_diferente= len(df_diferente)
print(tamanho_df_diferente)
```

```python
df_diferente = df_diferente.sort_values(by='category', ascending=False)
df_diferente.head()
```

```python
def quantidaCategoria(dataFrameComCategoria):
    df = pd.DataFrame(columns=['QTD', 'Categoria','PorcentagemTotal','QTD_Massa_Teste', 'Porcentagem_Massa_teste'])
    arrayCategoriaUnica = dataFrameComCategoria.category.unique()
    for x in (arrayCategoriaUnica):  
       DataCountCategoria = dataFrameComCategoria[dataFrameComCategoria.category.str.contains(x)]
       qtdCategoria = DataCountCategoria.category.count()
       df = df.append({'QTD':qtdCategoria ,'Categoria' :x,'PorcentagemTotal':(qtdCategoria*100)/tamanho_df_diferente} , ignore_index=True) 
    df=df.sort_values(by=['QTD'])
    return df

df_quantidade = quantidaCategoria(df_diferente)
df_quantidade.head()
```

```python
df_quantidade=df_quantidade.sort_values(by='QTD', ascending=False)
ax = df_quantidade.plot.bar(x='Categoria', y='QTD', rot=0)
```

```python
df_treino = pd.read_csv("limpoNaoClassificado_java.csv", sep="__label__",engine='python',header=1,names=["TITLE", "Categoria"])
df_treino.head()
```

```python
def quantidaCategoria(dataTesteCompleto,df_quantidade):
    arrayCategoriaUnica = df_quantidade.Categoria.unique()   
    for x in (arrayCategoriaUnica):  
       DataCountCategoria = dataTesteCompleto[dataTesteCompleto.Categoria.str.contains(x)]
       qtdCategoria = DataCountCategoria.Categoria.count()
       df_quantidade.loc[df_quantidade.Categoria.str.contains(x),'QTD_Massa_Teste']=qtdCategoria 
       df_quantidade.loc[df_quantidade.Categoria.str.contains(x),'Porcentagem_Massa_teste']=(qtdCategoria*100)/len(dataTesteCompleto)  
    return df_quantidade


df_quantidade_treino = quantidaCategoria(df_treino,df_quantidade)
```
